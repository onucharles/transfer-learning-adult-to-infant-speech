```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import os
import scipy
import librosa.display
from IPython.display import Audio
```

```python
#noise, noise_sr = librosa.load('/Users/jon/Desktop/gaussian_0_1noise.wav')
in_len = 90000

noise, noise_sr = librosa.load('/Users/jon/Downloads/train/dog_bark/4.wav')
sample, sample_sr = librosa.load('./p229_002.wav')
noise = librosa.resample(noise, noise_sr, target_sr=sample_sr)
print(len(noise))
noise = np.pad(noise, (0, max(0, in_len - len(noise))), "constant") 
sample = np.pad(sample, (0, max(0, in_len - len(sample))), "constant") 
mix = 1 * noise[:in_len] + 1 * sample[:in_len]
Audio(mix, rate=sample_sr)
```
